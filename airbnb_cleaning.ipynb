# Joining Files and Reducing Data to under 1 GB

```python
import pandas as pd
```

Read in New York listings, LA listings, and airbnb reviews.

```python
ny_listings = pd.read_csv('NY_Listings.csv', encoding = 'latin-1')
```

```python
la_listings = pd.read_csv('LA_Listings.csv', encoding = 'latin-1')
```

```python
airbnb_reviews = pd.read_csv('airbnb-reviews.csv', sep = ';')
```

Remove unnecessary columns

```python
ny_listings = ny_listings.drop(columns = ['Street', 'latitude', 'longitude', 'Minimum nights','Maximum nights', 'Availability 365', 'Calendar last scraped', 'Last Review Date'])
```

```python
la_listings = la_listings.drop(columns = ['Street', 'latitude', 'longitude', 'Minimum nights','Maximum nights', 'Availability 365', 'Calendar last scraped', 'Last Review Date'])
```

Inspeact the size of each dataframe

```python
ny_listings.shape
```

```python
la_listings.shape
```

```python
airbnb_reviews.shape
```

Change column name to 'listing_id'. This will make merging with airbnb_reviews easier.

```python
ny_listings = ny_listings.rename(columns = {'Listing ID':'listing_id'})
```

```python
la_listings = la_listings.rename(columns = {'Listing ID':'listing_id'})
```

```python
airbnb_reviews.head()
```

```python
len(airbnb_reviews.listing_id.unique())
```

```python
#Some reviews didn't add a comment. We can get rid of that in Databricks
airbnb_reviews.isnull().sum(axis=0)
```

Merge New York listings with reviews

```python
ny = pd.merge(ny_listings, airbnb_reviews, on = "listing_id")
```

```python
ny.shape
```

```python
ny
```

Merge LA listings with reviews

```python
la = pd.merge(la_listings, airbnb_reviews, on = "listing_id")
```

```python
la.shape
```

```python
la
```

Put both dataframes together using append

```python
airbnb = ny.append(la)
```

```python
airbnb.shape
```

Save as csv and now we can work with this dataset in Databricks

```python
airbnb.to_csv('airbnb.csv', index = False)
```

```python
from langdetect import detect
```

```python
def detect_en(text):
    try:
        return detect(text) == 'en'
    except:
        return False
```
